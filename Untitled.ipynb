```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
semi_a = 0.6
semi_b = 0.3
ecc = np.sqrt(1-(semi_b/semi_a)**2)
r = semi_a 

theta = np.linspace(0, 2*np.pi, 100)
rr = semi_a*(1+ecc*np.cos(theta))
```

```python
plt.plot(rr*np.cos(theta), rr*np.sin(theta))
```
